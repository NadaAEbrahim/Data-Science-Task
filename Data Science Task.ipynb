```python
# Importing libaraies
import numpy as np  
import pandas as pd  
import matplotlib.pyplot as plt
```

# Import dataset

```python
# load the dataset
df = pd.read_csv('drinkMenu.csv')
```

```python
# Data Exploration
df.head()
```

# Data preparation

1 - Remove duplicates

```python
# drop_duplicates
df.drop_duplicates(inplace=True)
```

```python
#Invalid data
duplicate = df.duplicated()
print(duplicate.sum())
```

2 - Fill null values

```python
# check the null values
df.isnull().sum()
```

```python
# fill nan value with the mode 
df['Caffeine (mg)'].fillna(df['Caffeine (mg)'].mode()[0],inplace=True)
```

```python
# check the null values
df.isnull().sum()
```

3 - Drop unnecessary Columns

```python
# drop 'Beverage_category' column
df=df.drop(["Beverage_category"],axis=1)      
```

```python
df.head()
```

# Data Visualisations

1- Which drink has the highest calories?

```python
fig, ax = plt.subplots(figsize =(10, 10))
calories = df['Calories']
drinks = df['Beverage']
# Horizontal Bar Plot
ax.barh(drinks, calories)

plt.xlabel("calories")
plt.ylabel("drink")
plt.title("Which drink has the highest calories")
plt.show
```

we find that ( White Chocolate Mocha (Without Whipped Cream) ) has the highest calories

2- Highest Sugar Drink ?

```python
fig, ax = plt.subplots(figsize =(15, 10))
sugars = df[' Sugars (g)']
drinks = df['Beverage']
# Horizontal Bar Plot
ax.barh(drinks, sugars)

plt.xlabel("sugars")
plt.ylabel("drink")
plt.title("Highest Sugar Drink ?")
plt.show
```

we find that ( Java Chip (Without Whipped Cream) ) has the highest sugar
